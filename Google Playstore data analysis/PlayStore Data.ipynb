## Filling up missing values

```python
# mode of categorical data that contains nan values 
print('-------------------------------------')
print(playstore_data['Type'].mode())
print('-------------------------------------')
print(playstore_data['Current Ver'].mode())
print('-------------------------------------')
print(playstore_data['Android Ver'].mode())
```

```python
#Filling the missing categorical values with modes 
playstore_data['Type'].fillna(str(playstore_data['Type'].mode().values[0]),inplace=True)
playstore_data['Current Ver'].fillna(str(playstore_data['Current Ver'].mode().values[0]),inplace=True)
playstore_data['Android Ver'].fillna(str(playstore_data['Android Ver'].mode().values[0]),inplace=True)
```

```python
#Counting the null values in each column
playstore_data.isnull().sum()
```

```python
playstore_data.head()
```


Converting the price, rating and review into numerical column 

```python
# Converting the price to float
playstore_data['Price'] = playstore_data['Price'].apply(lambda x: str(x).replace('$','') if '$' in str(x) else str(x))
playstore_data['Price'] = playstore_data['Price'].apply(lambda x: str(x).replace(',','') if '$' in str(x) else str(x))
playstore_data['Price'] = playstore_data['Price'].apply(lambda x: float(x))
```

```python
# Converting the installations to float 
playstore_data['Installs'] = playstore_data['Installs'].apply(lambda x: str(x).replace('+','') if '+' in str(x) else str(x))
playstore_data['Installs'] = playstore_data['Installs'].apply(lambda x: str(x).replace(',','') if ',' in str(x) else str(x))
playstore_data['Installs'] = playstore_data['Installs'].apply(lambda x: float(x))
```

```python
# Converting the Reviews to float
playstore_data['Reviews'] = playstore_data['Reviews'].apply(lambda x: str(x).replace('$','') if '$' in str(x) else str(x))
playstore_data['Reviews'] = playstore_data['Reviews'].apply(lambda x: str(x).replace(',','') if '$' in str(x) else str(x))
playstore_data['Reviews'] = playstore_data['Reviews'].apply(lambda x: float(x))
```

## Merging the two dataframes to do combined data analysis 

```python
# Using merge to merge two columns 
new_df = playstore_data.merge(user_reviews, on="App")
new_df.head()
```

```python
category_sentiment = new_df.groupby(['Category','Sentiment']).size().reset_index(name='Sentiment Count')
category_sentiment = category_sentiment.pivot('Category', 'Sentiment', 'Sentiment Count').fillna(0).reset_index().set_index('Category')
```
